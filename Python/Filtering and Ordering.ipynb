# Filtering and Ordering

```python
import pandas as pd 
```

```python
df = pd.read_csv('world_population.csv')
```

```python
df
```

```python
df[df['Rank'] <= 10]
```

```python
specific_countries = ['India', 'China']
```

```python
df[df['Country'].isin(specific_countries)]
```

```python
df[df['Country'].str.contains('America')]
```

```python
df2 = df.set_index('Country')
df2
```

```python
df2.filter(items = ['Continent', 'CCA3'], axis=1)
```

```python
df2.filter(items = ['India'], axis=0)
```

```python
df2.filter(items=['China'], axis=0)
```

```python
df2.filter(like='Sri', axis=0)
```

```python
df2.iloc[2]
```

```python
df[df['Rank'] < 5].sort_values(by=['Rank', 'Continent'])
```
