# Pandas Visualizations

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Ice Cream Ratings.csv')
```

```python
df
```

```python
df = df.set_index('Date')
df
```

```python
df.plot()
```

```python
df.plot(kind ='line', title='Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot(kind='bar')
```

```python
df.plot(kind='bar', stacked=True)
```

```python
df['Flavor Rating'].plot(kind='bar', stacked=True)
```

```python
df.plot.barh(stacked=True)
```

```python
df.plot.scatter(x='Texture Rating', y = 'Overall Rating', s=500, c='Red')
```

```python
df.plot.hist(bins=20)
```

```python
df.boxplot()
```

```python
df.plot.area()
```

```python
df.plot.pie(y='Flavor Rating')
```
